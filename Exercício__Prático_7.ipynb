## 4. KPIs globais da operação

Com base nas “4 Métricas de Ouro do Varejo” apresentadas na aula, calcule os seguintes KPIs globais da base inteira: fileciteturn5file0

- **Receita Total**
- **Lucro Total**
- **Margem de Lucro (%)**
- **Ticket Médio**

### Orientação

Receita Total = soma da coluna receita

Lucro Total = soma da coluna lucro

Margem de Lucro (%) = (Lucro Total / Receita Total) × 100

Ticket Médio = média da coluna receita

```python
# Calcule aqui os 4 KPIs globais

receita_total = df['receita'].sum()

lucro_total = df['lucro'].sum()

margem = (lucro_total / receita_total) * 100

ticket_medio = df['receita'].mean()

print(receita_total, lucro_total, margem, ticket_medio)
```

## 5. Métricas derivadas

A aula mostra que **somar valores não basta**.  
Precisamos criar métricas derivadas para avaliar eficiência e comportamento. fileciteturn5file0

### Tarefa
Crie, quando fizer sentido:
- uma coluna de **margem_lucro** (`lucro / receita`)
- uma lógica para **ticket_medio** da operação ou por grupo

### Reflexão
Explique:
1. Por que receita alta não significa automaticamente melhor desempenho?
2. Em que situação a margem ajuda mais do que a receita bruta?


```python
# Crie aqui as métricas derivadas

df['margem_lucro'] = df['lucro'] / df['receita']

ticket_medio_global = df['receita'].sum() / df['pedido_id'].nunique()

# um exemplo de agrupamento por canal
ticket_medio_canal = df.groupby('canal_venda').apply(
    lambda x: x['receita'].sum() / x['pedido_id'].nunique()
)

print(df.head())
print(ticket_medio_global)
print(ticket_medio_canal)
```

Uma receita elevada não significa necessariamente um bom desempenho, pois pode estar acompanhada de custos igualmente altos. Dessa forma, mesmo vendendo bastante, a empresa pode ter um lucro baixo. Sem analisar indicadores como lucro e margem, não dá para avaliar se o negócio é realmente eficiente.

A margem se torna mais importante quando o objetivo é analisar a rentabilidade. Por exemplo, ao comparar produtos ou canais, um pode ter menor volume de vendas, mas gerar maior lucro proporcional. Nesses casos, a margem ajuda a identificar onde vale mais a pena investir.

## 7. Análise por canal

### Pergunta de negócio
**Qual canal de venda é o mais eficiente?**

### Tarefa
Crie uma tabela agregada por `canal_venda` contendo, no mínimo:
- Receita Total
- Lucro Total
- Margem Média ou Margem Calculada
- Quantidade de pedidos
- Ticket Médio

### Depois responda
1. Qual canal gera mais receita?
2. Qual canal gera maior eficiência?
3. Se você fosse gestor, priorizaria volume ou margem?


```python
# Gere aqui a tabela agregada por canal_venda

tabela_canal = df.groupby('canal_venda').agg({
    'receita': 'sum',
    'lucro': 'sum',
    'pedido_id': 'nunique'
}).rename(columns={
    'receita': 'Receita Total',
    'lucro': 'Lucro Total',
    'pedido_id': 'Qtd Pedidos'
})

tabela_canal['Margem (%)'] = (tabela_canal['Lucro Total'] / tabela_canal['Receita Total']) * 100

tabela_canal['Ticket Médio'] = tabela_canal['Receita Total'] / tabela_canal['Qtd Pedidos']

print(tabela_canal)
```

O canal Marketplace geralmente apresenta a maior receita total, o que indica um volume mais alto de vendas.

Já analisando a margem, o canal Online se mostra mais eficiente, pois gera melhor retorno em relação às vendas realizadas.

Se eu fosse gestor, daria prioridade à margem, pois ela permite avaliar melhor a rentabilidade e a eficiência do negócio, contribuindo para um crescimento mais sustentável ao longo do tempo.

### Insight obrigatório

O canal Marketplace contribui bastante para o aumento do volume de vendas e da receita, porém isso não garante maior eficiência. Por outro lado, canais com margens mais altas indicam melhor rentabilidade por venda. Assim, a estratégia mais adequada é priorizar a margem, buscando maximizar o lucro de forma sustentável ao longo do tempo.

## 9. Análise por região (UF)

### Pergunta de negócio
**Quais UFs concentram a operação?**

### Tarefa
Monte uma tabela por `uf` com:
- Receita Total
- Lucro Total
- Participação percentual da receita
- Ticket Médio

### Depois responda
1. Quais UFs concentram maior receita?

3. Se fosse necessário expandir ou reforçar logística, por onde você começaria?


```python
# Gere aqui a análise por UF

receita_total_geral = df['receita'].sum()

# Agrupamento por UF
tabela_uf = df.groupby('uf').agg({
    'receita': 'sum',
    'lucro': 'sum',
    'pedido_id': 'nunique'
}).rename(columns={
    'receita': 'Receita Total',
    'lucro': 'Lucro Total',
    'pedido_id': 'Qtd Pedidos'
})

tabela_uf['Participação (%)'] = (
    tabela_uf['Receita Total'] / receita_total_geral
) * 100

tabela_uf['Ticket Médio'] = (
    tabela_uf['Receita Total'] / tabela_uf['Qtd Pedidos']
)

tabela_uf = tabela_uf.sort_values(by='Receita Total', ascending=False)

print(tabela_uf)
```

Estados como SP, RJ e PE concentram a maior parte da receita, o que indica um volume mais elevado de vendas nessas regiões.

Sim, existe uma concentração significativa no Sudeste, região que costuma ter maior densidade populacional e poder de consumo.

Para melhorar a logística, priorizaria inicialmente SP e RJ, por serem os estados com maior geração de receita. Já para expansão, o foco seria em regiões que ainda apresentam menor receita e lucro.

## 10. O motor da análise: agregações (groupby)

Nos slides, a lógica central é:
- `receita -> sum`
- `margem -> mean` ou cálculo derivado
- `produto -> count`
fileciteturn5file0

### Tarefa
Escolha **duas dimensões diferentes** e construa tabelas agregadas mostrando que você entendeu essa lógica.

Exemplos:
- por canal
- por categoria
- por UF
- por mês


```python
# Demonstre aqui o uso de groupby com diferentes agregações

tabela_categoria = df.groupby('categoria').agg({
    'receita': 'sum',
    'lucro': 'sum',
    'pedido_id': 'count'
})

# aqui é a margem derivada
tabela_categoria['margem'] = tabela_categoria['lucro'] / tabela_categoria['receita']

print(tabela_categoria)

# agrupamento de uf
tabela_uf = df.groupby('uf').agg({
    'receita': 'sum',
    'lucro': 'sum',
    'pedido_id': 'count'
})

# margem derivada
tabela_uf['margem'] = tabela_uf['lucro'] / tabela_uf['receita']

print(tabela_uf)
```

## 12. Análise temporal

A aula reforça que sem tempo a análise fica estática. fileciteturn5file0

### Tarefa
1. Converta `data_venda` para data, se necessário
2. Crie colunas auxiliares como:
   - mês
   - trimestre (opcional)
3. Gere uma tabela agregada por mês com:
   - receita
   - lucro
   - ticket médio ou margem

### Depois responda
1. O lucro parece crescer ao longo do tempo?
2. O último trimestre foi melhor que o anterior?
3. Quais meses concentram as vendas de fim de ano?


```python
# Faça aqui a análise temporal mensal

# Conversão
df['data_venda'] = pd.to_datetime(df['data_venda'])

# aqui são as colunas auxiliares
df['mes'] = df['data_venda'].dt.to_period('M')
df['trimestre'] = df['data_venda'].dt.to_period('Q')

tabela_mes = df.groupby('mes').agg({
    'receita': 'sum',
    'lucro': 'sum',
    'pedido_id': 'nunique'
}).rename(columns={
    'receita': 'Receita',
    'lucro': 'Lucro',
    'pedido_id': 'Qtd Pedidos'
})

# Fazendo a margem e ticket médio
tabela_mes['Margem (%)'] = (tabela_mes['Lucro'] / tabela_mes['Receita']) * 100
tabela_mes['Ticket Médio'] = tabela_mes['Receita'] / tabela_mes['Qtd Pedidos']

print(tabela_mes)
```

Não é possível observar um crescimento contínuo. O lucro varia ao longo dos meses, o que indica um comportamento mais sazonal, sem uma tendência consistente de alta.

O último trimestre, principalmente nos meses de novembro e dezembro, costuma apresentar desempenho melhor em relação ao trimestre anterior, impulsionado por datas comerciais.

Novembro e dezembro concentram os maiores picos de receita e lucro, provavelmente devido à Black Friday e ao Natal, evidenciando a sazonalidade do negócio.

```python
# Desafio extra opcional

receita_total = df['receita'].sum()
lucro_total = df['lucro'].sum()
margem = lucro_total / receita_total
ticket_medio = df['receita'].mean()



# final
tabela_global = pd.DataFrame({
    'Indicador': ['Receita Total', 'Lucro Total', 'Margem', 'Ticket Médio'],
    'Valor': [receita_total, lucro_total, margem, ticket_medio]
})

#pra remover a notação cientifica
tabela_global['Valor'] = [
    f"R$ {receita_total:,.2f}",
    f"R$ {lucro_total:,.2f}",
    f"{margem:.2%}",
    f"R$ {ticket_medio:,.2f}"
]

print(tabela_global)

# por canal
tabela_canal = df.groupby('canal_venda').agg({
    'receita': 'sum',
    'lucro': 'sum',
    'pedido_id': 'nunique'
}).rename(columns={
    'receita': 'Receita Total',
    'lucro': 'Lucro Total',
    'pedido_id': 'Qtd Pedidos'
})

# KPIs derivados
tabela_canal['Margem'] = tabela_canal['Lucro Total'] / tabela_canal['Receita Total']
tabela_canal['Ticket Médio'] = tabela_canal['Receita Total'] / tabela_canal['Qtd Pedidos']

print(tabela_canal)
```

Sim, o KPI global acaba escondendo diferenças importantes entre os canais. Apesar da margem geral ser em torno de 28%, esse valor mistura canais com comportamentos bem diferentes. O Marketplace, por exemplo, tem grande peso na receita, mas pode ter margens menores, enquanto canais como Online ou Loja Física são mais eficientes, mesmo vendendo menos.

Isso mostra que analisar só o KPI geral pode levar a decisões erradas, como focar apenas no canal com maior receita. Quando se observa por canal, fica claro que alguns são mais rentáveis e importantes para sustentar o lucro. Portanto, a análise detalhada é essencial para decisões mais estratégicas.
